```python
# import Dependencies
import pandas as pd
import os
```

```python
# Reading the data
load_data = os.path.join("vine_table_data.csv")
vine_data_df = pd.read_csv(load_data, encoding="ISO-8859-1")
```

```python
# Displaying the Data
vine_data_df
```

```python
# Checking Data for vine == "Y" 
vine_data_df[vine_data_df["vine"] == "Y"]
```

```python
# Total number of votes greater than or equal to 20
total_votes_filtered_df = vine_data_df[vine_data_df["total_votes"] >= 20]
total_votes_filtered_df
```

```python
# Number of votes greater than or equal to 50%
votes_greater_df = total_votes_filtered_df[(total_votes_filtered_df["helpful_votes"]/total_votes_filtered_df["total_votes"]) >= 0.5]
votes_greater_df
```

```python
# Review was written as part of the Vine program (paid), vine == 'Y'.
vine_paid_df = votes_greater_df[votes_greater_df["vine"] == "Y"]
vine_paid_df
```

```python
# Review was written not as a part of the Vine program (unpaid), vine == 'N'.
vine_unpaid_df = votes_greater_df[votes_greater_df["vine"] == "N"]
vine_unpaid_df
```

```python
# The number of reviews for paid program
number_reviews_paid = vine_paid_df["review_id"].count()
print(f"The number of reviews for paid program are {number_reviews_paid}")
```

```python
#The number of reviews for un-paid program
number_reviews_unpaid = vine_unpaid_df["review_id"].count()
print(f"The number of reviews for un-paid program are {number_reviews_unpaid}")
```

```python
# Calulate the number of 5 star reviews for unpaid program
number_5star_unpaid = vine_unpaid_df[vine_unpaid_df["star_rating"] == 5]
count_unpaid = number_5star_unpaid["star_rating"].count()
print(f"The number of 5 star reviews for unpaid program are {count_unpaid}")
```

```python
# Calculate the percentage of 5 star reviews for unpaid program
percent_unpaid = round(count_unpaid/number_reviews_unpaid *100, 2)
print(f"The percentage of 5 star reviews for unpaid program is {percent_unpaid}%")
```
